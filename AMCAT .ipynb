# Univariate Analysis

### For Numerical Columns:

```python
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
plt.boxplot(df['salary'])
plt.title("Box plot of salary")
```

### we should remove the outliers from salary column

```python
Q1 = df['salary'].quantile(0.25)
Q3 = df['salary'].quantile(0.75)
IQR = Q3 - Q1

# Determine the bounds
lower_bound = Q1 - 1.5 * IQR
upper_bound = Q3 + 1.5 * IQR

# Filter the DataFrame
df_filtered = df[(df['salary'] >= lower_bound) & (df['salary'] <= upper_bound)]

# Check the shape of the new DataFrame
print(f'Original DataFrame shape: {df.shape}')
print(f'Filtered DataFrame shape: {df_filtered.shape}')
```

```python
plt.boxplot(df_filtered['salary'])
plt.title('Box plot of filtered salary column')
```

Observations: The box plot shows that the median salary is around 300,000, with the interquartile range between 200,000 and 400,000.
There are a few outliers above 600,000, indicating a small number of employees earning significantly more than the majority.
The distribution appears relatively symmetric, but with a higher spread of salaries in the upper range.

```python
sns.histplot(df['salary'],kde = True)
plt.grid()
plt.title("Salary distribution")
plt.show()
```

observation:

## Research Question

### Times of India article dated Jan 18, 2019 states that â€œAfter doing your Computer Science Engineering if you take up jobs as a Programming Analyst, Software Engineer, Hardware Engineer and Associate Engineer you can earn up to 2.5-3 lakhs as a fresh graduate

```python
cs_engineers = df[df['computerscience'] != -1]  # Filter rows with Computer Science background
avg_salary = cs_engineers['salary'].mean()
print(f"Average Salary for CS Engineers: {avg_salary}")
```

```python
from scipy import stats
```

```python
cs_graduates = df[df['computerscience'] != -1]  

roles = ['Programming Analyst', 'Software Engineer', 'Hardware Engineer', 'Associate Engineer']
cs_graduates_roles = cs_graduates[cs_graduates['designation'].isin(roles)]


cs_fresh_graduates = cs_graduates_roles[pd.to_datetime(cs_graduates_roles['doj']).dt.year >= 2017]  
mean_salary = cs_fresh_graduates['salary'].mean()

t_stat, p_value = stats.ttest_1samp(cs_fresh_graduates['salary'], 275000)  # comparing with midpoint (2.75L)

print(f'Mean salary of fresh graduates: {mean_salary}')
print(f'T-statistic: {t_stat}, P-value: {p_value}')

if p_value < 0.05:
    print("Reject null hypothesis: Fresh graduates' salaries differ from the expected range (2.5-3 lakhs).")
else:
    print("Fail to reject null hypothesis: Fresh graduates' salaries fall within the expected range.")
```

## Is there any relationship between Gender and specialization?

```python
from scipy.stats import chi2_contingency

contingency_table = pd.crosstab(df['gender'], df[['computerscience', 'mechanicalengg', 'electricalengg', 'telecomengg', 'civilengg']].apply(lambda row: row.idxmax(), axis=1))
chi2, p, dof, expected = chi2_contingency(contingency_table)

print(f'Chi-square test statistic: {chi2}, P-value: {p}')

if p < 0.05:
    print("Reject null hypothesis: There is a relationship between gender and specialization.")
else:
    print("Fail to reject null hypothesis: No significant relationship between gender and specialization.")
    
contingency_table.plot(kind='bar', stacked=True)
plt.title('Gender vs Specialization')
plt.ylabel('Count')
plt.show()
```
